## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
import time
for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 20000);")
```

```python
from datetime import date

print(date.today().strftime('%Y/%m/%d'))
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[:10] == date.today().strftime('%Y/%m/%d'):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
import datetime
dt1 = datetime.datetime.now()
dt2 = dt1 - datetime.timedelta(hours = 2)
for d in soup.find(class_="part_list_2").find_all('h3'):
    dttime = d.find(class_="date").text
    dt = datetime.datetime(int(dttime[:4]),int(dttime[5:7]),int(dttime[8:10]),int(dttime[11:13]),int(dttime[14:16]))
    if dt1 >= dt and dt >= dt2:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
browser.quit();
```

```python
import datetime

today = datetime.date.today()
tdago1 = datetime.datetime(today.year,today.month,today.day,15,0) - datetime.timedelta(days = 3)
tdago2 = tdago1 + datetime.timedelta(hours = 2)

from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value(str(tdago1.year))
selectSite = Select(browser.find_element_by_id("selM"))
selectSite.select_by_value(str(tdago1.month))
selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value(str(tdago1.day))
browser.find_element_by_id('button').click()

for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 20000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    dttime = d.find(class_="date").text
    dt = datetime.datetime(int(dttime[:4]),int(dttime[5:7]),int(dttime[8:10]),int(dttime[11:13]),int(dttime[14:16]))
    if tdago2 >= dt and dt >= tdago1:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
browser.quit();
```
